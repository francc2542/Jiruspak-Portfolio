**Attribute Information**
- **nvoice_no :** Invoice number. Nominal. A combination of the letter 'I' and a 6-digit integer uniquely assigned to each operation.
- **customer_id :**  Customer number. Nominal. A combination of the letter 'C' and a 6-digit integer uniquely assigned to each operation.
- **gender :** String variable of the customer's gender.
- **age :** Positive Integer variable of the customers age.
- **category :** String variable of the category of the purchased product.
- **quantity :** The quantities of each product (item) per transaction. Numeric.
- **price :** Unit price. Numeric. Product price per unit in Turkish Liras (TL).
- **payment_method :** String variable of the payment method (cash, credit card or debit card) used for the transaction.
- **invoice_date :** Invoice date. The day when a transaction was generated.
- **shopping_mall :** String variable of the name of the shopping mall where the transaction was made.

```python
# Import libraries and the .csv file.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('customer_shopping_data.csv', encoding = 'utf-8')
df
```

```python
# Shape of dataframe.
df.shape
```

```python
# Preview top 5 rows.
df.head()
```

```python
# Preview bottom 5 rows.
df.tail()
```

```python
# See data frame information using .info()
df.info()
```

```python
# Convert the data type of the 'invoice_date' object to datetime in the original dataframe.
df['invoice_date'] = pd.to_datetime(df['invoice_date'], format='%d/%m/%Y')
df.sample(5)
```

```python
# Calculate the total price.
df['Total_price'] = df['quantity'] * df['price']
df.sample(5)
```

```python
# Additionally, we create new columns for the year and month, which can be useful for further analysis.
df['invoice_month'] = df['invoice_date'].dt.month
df['invoice_year'] = df['invoice_date'].dt.year
df.sample(5)
```

```python
df.info()
```

```python
# Check missing values in each column
df.isnull().sum()
```

```python
# Summary stats.
df[['Total_price', 'quantity', 'price', 'Total_price']].describe()
```

```python
# Create a line graph to observe the monthly income trends.
year_month = df.groupby(['invoice_year', 'invoice_month'])['Total_price'].sum()
plt.figure(figsize=(10, 8))
year_month.plot(marker = "o")
plt.title('The trend of monthly sales', size = 20)
plt.xlabel('Year, Month', size = 10)
plt.ylabel('Total Sales', size = 10)
plt.grid(True)
plt.show()
```

```python
# Prepare data to create a pie chart depicting income for each year.
year_price = df.groupby('invoice_year')['Total_price'].sum().reset_index()
year_price
```

```python
# Create a pie chart illustrating income for each year.
exp = [0, 0, 0.1]

plt.pie(year_price['Total_price'], labels = year_price['invoice_year'], autopct = "%.2f%%", shadow = True, explode = exp)
plt.title('Annual income.', size = 15)
plt.show()
```

```python
# Explore age distribution.
ages = df['age'].value_counts().reset_index()
ages.sort_values('age')
```

```python
# Create a histogram graph showing the age groups of customers who made purchases.
plt.hist(df['age'], bins = 20)
plt.xlabel("Age", size = 10)
plt.ylabel("Count", size = 10)
plt.title("Frequency Distribution of Age", size = 20)
plt.grid(True)
plt.show()
```

```python
# Prepare data to create a bar chart illustrating the number of purchases in each product category, segmented by gender.
category = df.groupby(['category','gender'])['quantity'].sum().reset_index()
category
```

```python
# Create a bar chart showing the number of purchases in each product category, segmented by gender.
sns.barplot(data=category, x = 'category', y = 'quantity', hue='gender')
plt.xlabel('Categories', size = 10)
plt.ylabel('Qntity', size = 10)
plt.title('The number of orders in each product category.', size = 15)
plt.grid(True)
plt.xticks(rotation=50)
plt.show()
```

```python
# Prepare the data to create a bar chart showing the income in each product category, segmented by gender.
Total_price_category = df.groupby(['category','gender'])['Total_price'].sum().reset_index()
Total_price_category
```

```python
# Create a bar chart showing the income in each product category, segmented by gender.
sns.barplot(data=Total_price_category, x = 'category', y = 'Total_price', hue = 'gender')
plt.xlabel('Categories', size = 10)
plt.ylabel('Total Sales', size = 10)
plt.title('The total revenue for each product category.', size = 15)
plt.grid(True)
plt.xticks(rotation=50)
plt.show()
```

```python
# Prepare data to create a bar chart illustrating the income in each shopping mall.
shopping_mall_price = df.groupby('shopping_mall')['Total_price'].sum().reset_index()
shopping_mall_price
```

```python
# Create a bar chart to see which shopping mall has the highest income.
plt.barh(shopping_mall_price['shopping_mall'], shopping_mall_price['Total_price'])
plt.xlabel("Mall", size = 10)
plt.ylabel("Total Sales", size = 10)
plt.title("The total revenue for each shopping mall.", size = 20)
plt.show()
```

```python
# Prepare the data to create a bar chart showing the number of customers who use the services in each shopping mall, segmented by gender.
shopping_mall_counts = df.groupby(['shopping_mall','gender'])['gender'].value_counts().reset_index()
shopping_mall_counts
```

```python
# Create a bar chart to see which shopping mall has the highest number of customers using services, segmented by gender.
sns.barplot(data=shopping_mall_counts, x = 'shopping_mall', y = 'count', hue = 'gender')
plt.xlabel('Mall', size = 10)
plt.ylabel('Count', size = 10)
plt.title('The number of people who visited each shopping mall.', size = 15)
plt.grid(True)
plt.xticks(rotation=50)
plt.show()
```

```python
# Prepare the data to create a pie chart showing popular payment methods.
payment = df.groupby('payment_method')['payment_method'].value_counts().reset_index()
payment
```

```python
# Create a pie chart to visualize popular payment methods.
exp = [0, 0, 0.1]

plt.pie(payment['count'], labels = payment['payment_method'], autopct = "%.2f%%", shadow = True, explode = exp)
plt.title('Popular payment methods.', size = 15)
plt.show()
```
